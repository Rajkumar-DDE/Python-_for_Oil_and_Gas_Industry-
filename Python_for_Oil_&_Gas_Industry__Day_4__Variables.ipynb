# Rules For Variable Names

1.   Variable always start with a litter or an underscore
2.   Variable name only contains alphabets, numeric values and underscore
3.   Variable names never start with number
4.   Variable names are case sensitive



```python
# Applying Rule No 1 (Variable always start with a litter or an underscore)
```

```python
xyz= 345
```

```python
_xyz= 345
```

```python
print(xyz)
print(_xyz)
```

```python
print(xyz+_xyz)
```

```python
1xyz = 345
```

```python
_1xyz = 345
```

```python
1_xyz = 345
```

```python
# Applying Rule No 2 (Variable name only contains alphabets, numeric values and underscore)
```

```python
_cd34= 45678
```

```python
cd_34= 3456
```

```python
_cd23@ = 4567
```

```python
# Applying Rule No 3 (Variable names never start with number)
```

```python
rty = 34567
```

```python
_rty = 35467
```

```python
123rty = 4567
```

```python
# Applying Rule No 4 (Variable names are case sensitive)
```

```python
raj = 345
Raj = 567
```

```python
print(raj)
print(Raj)
```

```python
print(raj + Raj)
```
