# String Concatenation

**Add the multiple string into a single string**

```python
first = 'Bachelor of Technology'
middle = 'Petroleum Engineering'
last = 'Rajkumar'
```

```python
# No need to use "" or '' in variable names
```

```python
print(first + middle + last)
```

```python
# we are not get the space in this variable , # lets creat new variable
```

```python
first_2 = 'Bachelor of Technology'
middle_2 = 'Petroleum Engineering'
last_2 = 'Rajkumar'
```

```python
print(first_2 +' '+ middle_2 +' ' +last_2)
```

**NOTE: Only string can be added to a string**

```python
print(first + '2')
```

```python
print(first +  str(2))
```

```python
print(first +  True)
```

```python
print(first + str(True))
```
