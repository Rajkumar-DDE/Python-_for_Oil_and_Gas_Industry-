# STRING FORMATING

```python
field = "Mumbai offshore"
press = 3500 #psi
```

```python
#ugly method (Concatentaion)
print('Reservoir pressure is' + ' ' + press +  ' ' + 'psi')  # why come to this error?, Because the variable name (press) is an intager value convert to string value
```

```python
print('Reservoir pressure is' + ' ' + str(press) +  ' ' + 'psi') #you get output value #but you are using long code
```

```python
#now using ugly method like {}    #here you get output same, but using coding method is different #for ex python 3
print('Reservoir pressure is {} psi'.format(press))
```

```python
print(f'Reservoir pressure is {press} psi')  #you get same output but coding , Because dependon Python version #for ex python 3.12
```

```python
# output required: Reservoir pressure in Mumbai offshore is 3500 psi  # it's very complex medhod
print('Reservoir pressure in' + ' ' + field + ' '+ 'is'+ '  ' + str(press) +  ' ' + 'psi')
```

```python
# Now using python 3 method
print('Reservoir pressure in {} is {} psi'.format(field, press)) # order is improtant
```

```python
# Now i change the order,see the output..
print('Reservoir pressure in {} is {} psi'.format(press, field))
```

```python
#Now i creat temporary variable here.. # here order is not important becaus temporary variable given the output
print('Reservoir pressure in {field_name} is {pressure} psi'.format(pressure = press, field_name = field))

```

```python
# Now using python 3.12 method
print(f'Reservoir pressure in {field} is {press} psi')
```

```python
# We can also change pressure to any different value
# using pytho 3 method # in starting time i use pressue 3500, i want now 5000, how can I change it now? just see outpot value...
print('Reservoir pressure in {} is {} psi'.format(field, press+1500))
```

```python
# using pytho 3.12 method # in starting time i use pressue 3500, i want now 5000, how can I change it now? just see outpot value...
print(f'Reservoir pressure in {field} is {press+1500} psi')
```
